```python
export_2016 = pd.read_excel('Russia-2016.xlsx', sheet_name='Partner')
export_2017 = pd.read_excel('Russia-2017.xlsx', sheet_name='Partner')
export_2016.head()
```

```python
#Конкатенация таблиц
export = pd.concat([export_2016, export_2017])
#Случайные строки (sample)
export.sample(10)
```

```python
#Сводная таблица
export.pivot_table(index=['Year'], values='Export (US$ Thousand)', aggfunc='sum')
```

```python
exclude = ['Europe & Central Asia',' World', 'East Asia & Pacific', 'Middle East & North Africa']
#Фильтрация данных
export_filtered = export[export['Partner Name'] != 'Europe & Central Asia']
#Фильтрация данных
export_filtered = export[-export['Partner Name'].isin(exclude)]
```

```python
export_pivot = export_filtered.pivot_table(index=['Year', 'Partner Name'], values='Export (US$ Thousand)', aggfunc='sum')\
    .sort_values(['Year','Export (US$ Thousand)'], ascending=[True, False])
```
